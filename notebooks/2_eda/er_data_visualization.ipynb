```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/시계열 데이터 분석/팀플/시계열데이터/er_filtered_2020_2021.csv", encoding='utf-8-sig')

df["year_month"] = pd.to_datetime(df["year_month"])

plt.figure(figsize=(14, 6))
plt.plot(df["year_month"], df["visit_count"], marker='o', linestyle='-')

plt.title("서울시 월별 응급실 방문 횟수 (2020-2021)", fontsize=15)
plt.xlabel("연도-월", fontsize=12)
plt.ylabel("응급실 방문 횟수", fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
