```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/시계열 데이터 분석/팀플/시계열데이터/temp_processed_2020_2021.csv", encoding='utf-8-sig')

df["year_month"] = pd.to_datetime(df["year_month"])

plt.figure(figsize=(14, 6))
plt.plot(df["year_month"], df["TA_MAVG"], label="평균기온 (TA_MAVG)", marker='o')
plt.plot(df["year_month"], df["TMX_MAVG"], label="최고기온 (TMX_MAVG)", marker='^')
plt.plot(df["year_month"], df["TMN_MAVG"], label="최저기온 (TMN_MAVG)", marker='v')

plt.title("서울시 월별 기온 변화 (2020-2021)", fontsize=15)
plt.xlabel("연도-월", fontsize=12)
plt.ylabel("기온 (°C)", fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
